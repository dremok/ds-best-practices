```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
numerical_cols = ['LotFrontage', 'LotArea', 'OverallQual', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF','GrLivArea' , 'BsmtFullBath' , 'TotRmsAbvGrd', 'Fireplaces', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold']
```

```python
categorical_cols = ['MSSubClass', 'MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'Exterior1st', 'Foundation', 'SaleCondition']
```

```python
target_col = 'SalePrice'
```

```python
df = df[numerical_cols + categorical_cols + [target_col]]
```

```python
df.columns
```

```python
n_estimators = 100
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

numerical_transformer = SimpleImputer(strategy='mean')

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
])

model = RandomForestRegressor(n_estimators=100, random_state=0)

pipeline = Pipeline([
        ('preprocesser', preprocessor),
        ('model', model)
])
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

mae = -cross_val_score(pipeline, df.drop(target_col, axis=1), df[target_col], cv=5, scoring='neg_mean_absolute_error').mean()
```

```python
mae
```

```python
from catboost import CatBoostRegressor
from catboost import Pool, cv
```

```python
cv_dataset = Pool(data=df.drop(target_col, axis=1),
                  label=df[target_col],
                  cat_features=categorical_cols)
```

```python
params = {"iterations": 1000,
          "depth": 2,
          "loss_function": "MAE",
          "verbose": False}

scores = cv(cv_dataset,
            params,
            fold_count=5)
```

```python
scores['test-MAE-mean'][999]
```

```python
cat_train_mat = Pipeline([('preprocesser', preprocessor)]).fit_transform(df.drop(target_col, axis=1))
cat_train_df = pd.DataFrame.sparse.from_spmatrix(cat_train_mat)
```

```python
model = CatBoostRegressor(learning_rate=1, depth=6, loss_function='MAE')
fit_model = model.fit(cat_train_df, df[target_col])
```
